```python
import numpy as np
import itertools
from prospectFunctions import *
```

```python
data_grid_params = np.load('params.npy')
# [run_params['zred'], run_params['mass'], logz, dust, tage]
data_grid_spec = np.load('specs.npy')
data_grid_phot = np.load('phots.npy')
data_grid_break = np.load('breaks.npy')
```

```python
grid_ranges = {}
grid_ranges['logzsol'] = np.linspace(-1,.5,10)
grid_ranges['dust2'] = np.linspace(0,3,10)
grid_ranges['tage'] = np.linspace(7,9,10)

logzs = grid_ranges['logzsol']
dusts = grid_ranges['dust2']
tages = grid_ranges['tage']
```

```python
# Axes have been checked

for i_logz in range(len(logzs)):
    test = data_grid_break[i_logz,:,:,:]
    imshow(test, extent=[tages[0], tages[-1], dusts[-1], dusts[0]])
    colorbar(fraction=0.046, pad=0.04)
    title(f'logzsol={logzs[i_logz]:.2f}')

    xlabel('tage (Gyrs)')
    ylabel('dust2')
    
    show()
```

```python
# Axes have been checked

for i_dust in range(len(dusts)):
    test = data_grid_break[:,i_dust,:,:]
    imshow(test, extent=[tages[0], tages[-1], logzs[-1], logzs[0]])
    colorbar(fraction=0.046, pad=0.04)
    title(f'dust2={dusts[i_dust]:.2f}')

    xlabel('tage (Gyrs)')
    ylabel('logzsol')

    show()
```

```python
# Axes have been checked

for i_tage in range(len(tages)):
    test = data_grid_break[:,:5,i_tage,:]
    imshow(test, extent=[dusts[0], dusts[-1], logzs[-1], logzs[0]])
    colorbar(fraction=0.025, pad=0.04)
    title(f'tage={tages[i_tage]:.2f} (Gyrs)')

    xlabel('dust2')
    ylabel('logzsol')
    
    show()
```
