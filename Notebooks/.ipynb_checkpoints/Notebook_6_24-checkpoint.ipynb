```python
from matplotlib.pyplot import *
import numpy as np
from Functions_6_24 import *
```

```python
grid_ranges = {}
grid_ranges['tage'] = np.linspace(7,9,3)
grid_ranges['dust2'] = np.linspace(0,3,4)
grid_ranges['logzsol'] = np.linspace(-1,.5,4)

run_params = {}
run_params['zred'] = 3.548
run_params['mass'] = 1e8
run_params['logzsol'] = grid_ranges['logzsol'][0]
run_params['tage'] = grid_ranges['tage'][0]
run_params['dust2'] = grid_ranges['dust2'][0]
run_params['add_duste'] = False
run_params['zcontinuous'] = 1
```

```python
obs = getObs(**run_params)
```

```python
model = getModel(**run_params)
```

```python
# DOUBLE CHECK WHEN PLOTTING ###########################################################################

# Make sure logzsol prior maxi=0.19 does not impact inputted metallicity maxi=0.5
# Make sure dust2 prior maxi=2.0 does not impact inputted dust2 maxi=3.0
########################################################################################################
```

```python
sps = getSps(**run_params)
```

```python
init_spec, init_phot, init_mfrac = getTheta(model, obs, sps)
```

```python
title_text = ','.join(["{}={}".format(p,model.params[p][0]) for p in model.free_params])

a = 1.0 + model.params.get('zred', 0.0)

wphot = obs['phot_wave'] 

if obs['wavelength'] is None:
    wspec = sps.wavelengths
    wspec *= a
else:
    # Make obs['wavelength'] the range from xmin to xmax of interesting section of spectra (so we can truncate spectra that's stored)
    wspec = obs['wavelength']

xmin, xmax = np.min(wphot)*0.8, np.max(wphot)/0.8
temp = np.interp(np.linspace(xmin,xmax,10000), wspec, init_spec)
ymin, ymax = temp.min()*0.8, temp.max()/0.4

figure(figsize=(16,8))

loglog(wspec, init_spec, label='Model spectrum',
      lw=0.7, color='navy', alpha=0.7)
errorbar(wphot, init_phot, label='Model photometry',
         marker='s', markersize=10, alpha=0.8, ls='', lw=3,
         markerfacecolor='none', markeredgecolor='blue',
         markeredgewidth=3)
title(title_text)

xlabel('Wavelength [A]')
ylabel('Flux Density [maggies]')
xlim([xmin, xmax])
ylim([ymin, ymax])
legend(fontsize=20)
tight_layout()

show()
```
